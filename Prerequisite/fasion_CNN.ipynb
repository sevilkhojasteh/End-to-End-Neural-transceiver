```python
import tensorflow as tf
import numpy as np
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
model = tf.keras.Sequential([
    # the shape is expected to be (28, 28)
    tf.keras.Input(shape=(28, 28, 1)),
    # We generate 64 filters which are 3*3
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    # Flatten change the shape from square into single array
    tf.keras.layers.Flatten(),
    # This is hidden layers with 128 neurons with ReLu activation function
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    # This is the output layer which has the same amount of neuron as the #classes
    # With softmax activation function
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=5)
```
